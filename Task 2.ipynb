For task 2, we will use additional data called "Australian postcodes" to get the latitude and longitude of the suppliers and government agencies location. The additional data has lot of additional information on location, which can be utilized for additional analysis.

```python
import folium
import pandas as pd
from folium.plugins import FastMarkerCluster
```

```python
customer = pd.read_excel("2019-20-australian-government-contract-data.xlsx")
```

```python
post_codes = pd.read_csv("australian_postcodes.csv")
```

## For Suppliers

```python
def create_cluster(state):
    """Function to create cluster marker using folium FastMarkerCluster plugin
    
    :param state - state from where the supplier is. Is used to subgroup the main dataset, which then is merged with the 
                   additional dataset for latitude and longitude
    :return - map with the markers
    """
    customer["Supplier Suburb"] = customer["Supplier Suburb"].str.upper()
    supplier_details = customer.loc[:,["Supplier Suburb","Supplier Postcode","Supplier State"]]
    df = pd.merge(supplier_details, post_codes, left_on=["Supplier Suburb","Supplier Postcode"], right_on=["locality","postcode"])
    df = df[df["state"] == state]
    locations = list(zip(df['lat'].tolist(), df['long'].tolist()))

    office_map = folium.Map(location=[-25, 133], zoom_start=4)

    office_map.add_child(FastMarkerCluster(df[['lat', 'long']].values.tolist()))
        
    return office_map
```

```python
#Cluster map for Victoria State
create_cluster("VIC")
```

In Victoria, the supplier location is distributed all across but there are 4 main clusters near Melbourne CBD, Bendigo, Wodonga and Traralagon. So if the bank is looking to set up offices, they can set up at these locations and for Melbourne CBD, suburb like St Kilda or Hawthorn can be looked into.

## For Government Agencies

```python
def create_cluster_office():
    """Function to create cluster marker using folium FastMarkerCluster plugin
    
    :return - map with the markers
    """
    supplier_details = customer.loc[:,["Office Postcode"]]
    df = pd.merge(supplier_details, post_codes, left_on="Office Postcode", right_on="postcode")
    df.drop_duplicates(subset="Office Postcode", inplace=True)
    locations = list(zip(df['lat'].tolist(), df['long'].tolist()))

    office_map = folium.Map(location=[-25, 133], zoom_start=4)

    office_map.add_child(FastMarkerCluster(df[['lat', 'long']].values.tolist()))
        
    return office_map
```

```python
create_cluster_office()
```

For Government Agencies, their location is pretty much similar to supplier location, but in small numbers. So the office location decided for suppliers will also be for government agencies.
